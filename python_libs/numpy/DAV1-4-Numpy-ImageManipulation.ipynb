```python
import numpy as np
```

```python
a = np.arange(4)
a
```

```python
b = a.reshape(2, 2,)
b
```

```python
a[0] = 100
```

```python
a
```

```python
b
```

```python
np.shares_memory(a, b)
```

```python
# only possible when a and b are sharing memory because they have same data
# reshape doesnt lead to change in data hence data memory can be shared
```

```python
a = np.arange(4)
a
```

```python
b = a + 2
b
```

```python
np.shares_memory(a, b)
```

```python
# until now Numpy decided whether to share memory or not
# but what we want to havr the power
```

```python
# share memory --> shallow copy
# not shared memory --> deep copy
```

```python
a = np.arange(10)
b = a.view()
```

```python
a[0] = 100
a
```

```python
b
```

```python
a = np.arange(10)
b = a.copy()
a[0] = 100
print(np.shares_memory(a, b))
b
```

```python
# view - creates a new view but data is shared (shallow copy)
# copy - which creates a deep copy and data is not shared
```
