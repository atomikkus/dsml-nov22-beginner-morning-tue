### Fancy Indexing

```python
m1 = np.arange(12).reshape(3, 4)
m1
```

```python
m1 < 6 # mask or the condition
```

```python
m1[m1 < 6] # ideally it should return all numbers < 6
# The output of Fancy Indexing on a ND array will always be a 1D array
```
