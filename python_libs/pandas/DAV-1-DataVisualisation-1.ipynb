### Univariate Categorical Viz

```python
cat_counts = data["Genre"].value_counts()
cat_counts
```

- Distribution/frequency of different categories?
- %/proportion contribution of different categories?

```python
x = cat_counts.index
y = cat_counts.to_list()
x, y
```

```python
plt.bar(x, y)
```

 How can we handle overlapping labels?

 - Maybe decrease the font size (not preffered)
 - Increasre the size of the figure
 - Rotate the labels with certain angle

```python
plt.figure(figsize=(12,8))
plt.bar(x,y)
```

```python
plt.bar(x,y)
plt.xticks(fontsize=5)
plt.show()
```

```python
plt.bar(x,y)
plt.xticks(rotation = 90, fontsize=10)
plt.show()
```

```python
x = cat_counts.index
y = cat_counts.to_list() # we took the data in the sorted order
plt.figure(figsize=(5,3))
plt.bar(x,y, color="orange", width=0.5)
plt.xticks(rotation = 90, fontsize=7)
plt.yticks(fontsize=7)
plt.title("Games for different Genre")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
# figure(figsize), plot, properties, plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.countplot(data = data, x="Genre", color="orange", order=cat_counts.index[::-1])
plt.xticks(rotation = 90, fontsize=7)
plt.yticks(fontsize=7)
plt.title("Games per genre")
plt.show()
```

```python
sales_data = data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
region_sales = sales_data.T.sum(axis='columns') # list


plt.pie(region_sales,
       labels=region_sales.index,
       startangle=90,
       explode=(0.2,0,0,0))
plt.show()
```

```python
plt.pie(cat_counts.to_list(),
       labels=cat_counts.index,
       startangle=90)
plt.show()
```
