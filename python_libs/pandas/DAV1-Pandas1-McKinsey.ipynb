### Ops of columns

1. Accessing the column names - .columns, .keys()
2. Accesing the column(s) - [[  ]], dot notation (not recommended)
3. Unique categories - unique(), nunique(), value_counts()
4. Rename column(s) - .rename(single string or list, axis, inplace)
5. Delete - drop(single string or list, axis, inplace)
6. Add a column and Update - just use assignment operation, vectorisation possible
